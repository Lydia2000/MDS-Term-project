```python
from preprocessing_pca import fit_pca
from data_helpers import DataHolder

data = DataHolder()
train_datasets, test_datasets, expected_RUL_datasets = data.get()

feature = fit_pca(train_datasets, test_datasets, expected_RUL_datasets)
train_datasets, test_datasets, expected_RUL_datasets = feature.result_df()
```

```python
def Calculate_RUL(df):
    max_cycles = df.groupby('Unit Number')['Time (Cycles)'].max()
    merged = df.merge(max_cycles.to_frame(name='max_time_cycle'), left_on='Unit Number',right_index=True)
    merged["RUL"] = merged["max_time_cycle"] - merged['Time (Cycles)']
    merged = merged.drop("max_time_cycle", axis=1)
    return merged
```

```python
for i in range(4):
    train_datasets[i] = Calculate_RUL(train_datasets[i])
```
